### Practical considerations of transmission

## Line encoding
Needing to know line encoding isn't very common any more, but are still used in a few different situations, including automotive networks. (Note: line encoding is typically at baseband)

- **NRZ-L** - *non-return to zero* - 1s have a positive signal, 0s are inverted
- **NRZ-I** - Alternate on amplitude on 1s
- **Bipolar AMI** - 0s are 0, 1s alternate, so `[0,1,0,0,-1,1,0,0]` decodes to `[0,1,0,0,1,1,0,0]`
- **Pseudo-ternary** - the exact opposite ove Bipolar AMI - 1s are encoded as 0s, 0s alternate between 1 and -1
- **Manchester** - Fancy! Changes in the middle of each timestep. 0s become high-low, 1s are low-to-high
- **Diff Manchester** - Like Manchester, but each time there's a 1, how 1 and 0 are encoded swapps... *why?!*

### Routing
There are three goals when routing messages:
- Minimize Latency
- Maximize Reliability
- Minimize energy utilization...
These are usually combined int oa single cost (per link).

Dijstra and Bellman-Ford algorithms are usually used for computing shortest-path for each algorithm

We perform routing using a forwarding/routing table. (Each router has a table)
Each table basically has $(\text{IP}, \text{which link})$

Djikstra's algorithm (or Bellman-Ford) is used on a "small" collection of nodes, and figure out the local costs. Each one of these systems are referred to as **"Autonomous Systems"**.

**Autonomous Systems** are linked by **Gateways**, which are linked by *Border Gateway Protocols

#### BGP - Border Gateway Protocol
A protocol for exchanging routing and costs to different locations between (AS)

#### OSPF - Open Shortest Path First
An Interior Gateway Protocol (Used internally in an Autonomous System (AS))

"OSPF is perhaps the most widely used IGP [Interior gateway protocols] in large enterprise networks."

Performs Djikstra's algorithm, and also does **Flooding**.
Flooding is is the process of broadcasting information to *all* of your neighbors.
