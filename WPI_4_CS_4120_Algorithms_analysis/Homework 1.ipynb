```python
from random import shuffle

coins = [2]*80 + [1]
shuffle(coins)

# Returns a number > 0 iff left is heavier than right
# 0 iff equal, and a number < 0 iff right is heavier than left.
def balance(left, right):
    print('Used balance')
    return sum(left) - sum(right)

def find_lightest(start, length):
    if(length == 1):
        print("Coin {:} is light with weight {:}".format(start, coins[start]))
        return
    third = int(length/3) # NOTE: Rounds down
    scale_tipped_direction = balance(coins[start:start+third], coins[start+third:start+third*2])
    if scale_tipped_direction == 0: # light coin not weighed
        find_lightest(start+2*third, third)
    elif scale_tipped_direction > 0: # light coin on right
        find_lightest(start+third, third)
    else: # light coin on left
        find_lightest(start, third)
    
find_lightest(0, 81)
```

## Problem 4

Permutation: $_n P _k = \frac{n!}{(n-k)!}$
When $n = k$, $_n P _n = n!$
##### a
$n(n-1)/2$ is the maximum number of possible inversions in a list, and there's only one possible ordering that provides that number of inversions (strictly decreasing).
Thus the probability of $n(n-1)/2$ inversions is $1/n!$
##### b
There's only one possible ordering with zero inversions (strictly increasing sorted order).
Probability of 0 inversions is $ 1 / n!$
##### c
Exacty one inversion means that two consecutive items are swapped from being entirely ordered. There are $n - 1$ consecutive pairs, so:
Probability of 1 inversions is $ \frac{n-1}{n!}$

## Problem 5
$E(n) = 1$

When considered separately, each location has a chance of being in the correct posistion ($X_i$) of $1/n$, thus our expected value should be $E(n) = E(\Sigma(X_i))$

We can use the linearity of expectation to turn this into $\Sigma(E(X_i))=\Sigma(1/n) = n*1/n = 1$

Thus $E(n) = 1$
