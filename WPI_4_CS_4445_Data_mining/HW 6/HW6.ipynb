## DBSCAN clustering

```python
from sklearn.cluster import DBSCAN

from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)


#Compute clustering with DBSCAN. 

# Replace this block of code with your implementation of DBSCAN.
# you could use this code to check the correctness of your result.

db = DBSCAN(eps=0.1, min_samples=5).fit(X)

# cluster labels: noise points are labeled as -1
labels = db.labels_
# Number of Core Points 
core_samples = db.core_sample_indices_

#---------------------------------------
import numpy as np
NO_CLUSTER = -1

# These two variables are not thread safe...
core_points = []
clusters = 0

def next_cluster():
    global clusters
    clusters = clusters + 1
    return clusters

def distance(p1, p2):
    return np.linalg.norm(p1-p2)

# Returns the list of points that are neighbors to point, not including itself.
def find_neighbors(point_idx, X, epsilon):
    isNeighbor = lambda i: distance(X[i], X[point_idx]) <= epsilon and point_idx != i
    return filter(isNeighbor, range(len(X)))


def add_cluster(i, X, labels, epsilon, min_samples):
    open_list = [i]
    cluster = NO_CLUSTER
    while len(open_list) != 0:
        i = open_list.pop(0)
        if labels[i] == NO_CLUSTER: # We haven't already processed it
            neighbors = find_neighbors(i, X, epsilon)
            isCorePoint = len(neighbors) >= min_samples
            
            if cluster == NO_CLUSTER and isCorePoint:
                cluster = next_cluster()
            
            labels[i] = cluster
            if isCorePoint:
                core_points.append(X[i])
                if len(core_points) % 100 == 0:
                    print("Added %d core points" % len(core_points))
                open_list += neighbors

                
def db_scan(X, eps=0.1, min_samples=5):
    global clusters
    clusters = 0
    
    labels = np.zeros(len(X))
    labels[:] = NO_CLUSTER
    for i in range(len(X)):
        add_cluster(i, X, labels, eps, min_samples)
    
    return (labels, core_points)

labels, core_points = db_scan(X)
#-----------------------------------------


# Number of clusters 
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print('number of clusters: %d' % n_clusters)
n_core = len(core_samples)
print('number of core samples: %d' % n_core)
# Number of Noise Points
n_noise=sum(labels==-1)
print('number of noise samples: %d' %n_noise)
# Number of Boundary Points
n_boundary=len(labels) - n_core - n_noise
print('number of boundary samples: %d' %n_boundary)
```

# Problem 3: Spectral Clustering on Yelp Social Network (50 points)
In this question, please implement the spectral clustering algorithm. 
You could use the existing implementation to test the correctness of your code.

```python
import json
import numpy as np


# load users
users = []
for line in open('yelp_academic_dataset_user.json', 'r'):  
    users.append(json.loads(line))

# friend counts
friend_count = [len(u['friends']) for u in users]
# remove users with less than 200 friends
users = [x for (x,y) in zip(users, friend_count) if y>200]
number_users=len(users)

# user ids
ids = [u['user_id'] for u in users]

# create dictionary for id index
d= {}
for i in range(len(ids)):
    d[ids[i]]=i
```

## Extract the Social Network 

```python
# Compute Affinity matrix

A = np.zeros((number_users, number_users))

idx=0
for u in users:
    for f in u['friends']:
        if f in d:
            A[idx][d[f]]=1
    idx+=1

# symetric
A=(A+A.T)/2

# plot affinity matrix
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(A,cmap=plt.cm.gray)
plt.show()
```

## Spectral Clustering

```python
import numpy as np
from sklearn.cluster import spectral_clustering

#Compute clustering with Spectral Clustering.
#--------------------------------------
# Replace this block of code with your implementation of Spectral Clustering.
# you could use this code to check the correctness of your result.
labels = spectral_clustering(A, n_clusters=6,assign_labels='kmeans')
#--------------------------------------
n_clusters = 6

D = np.zeros_like(A)
for i in range(D.shape[0]):
    D[i,i] = sum(A[i])

L = D-A # Laplacian matrix

# Get the n_clusters lowest eigenvectors
eigVals, eigVecs = np.linalg.eig(L)
ev_list = zip( eigVals, eigVecs )
ev_list.sort(key=lambda pair: pair[0], reverse=False)
eigVals, eigVecs = zip(*ev_list)
```

```python
lowest_eigVecs = np.array(eigVecs)[:n_clusters]

# Convert a list<0 or 1> of friends into a six dimensional vector
def transform_point(friend_list):
    return np.array([np.dot(friend_list, v) for v in lowest_eigVecs])

X = np.array([transform_point(row) for row in A])

from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=n_clusters, n_init=10)
labels = k_means.fit_predict(X)
```

## plot the results

```python
import networkx as nx

G = nx.from_numpy_matrix(A)
pos = nx.spring_layout(G)
## put together a color map, one color for a category
colors = np.array([x for x in 'bgrcmy'])
## construct a list of colors then pass to node_color
nx.draw(G, pos,node_color=colors[labels].tolist(),node_size=35)
plt.show()
```
