## January 19

##### Displaying data (data -> viz feature) with Set theory terms
- Bijection: One visual attribute, one data attribute
- Surjection: Mutliple visual attributes to one data attribute
- Injection: One to one mapping, but not all data elements are mapped

##### Data attributes
- **Nominal:** ==, != (qualitative data - nonparametric)  Non-ordered and non-numeric (AKA categorical data)
- **Ordinal:** >, <, (qualitative data - nonparametric)  Ordered and non numeric (e.g. [1st, 3rd, 5th], ['G', 'PG', 'PG-13', 'R'])
- **Interval:** +, - Ordered numeric data, but can't divide it (e.g. ['Jan 12', 'Jan 20'], [17°, 44° (temperature)
- **Ratio:** *, / Ordered numeric, has a "true" 0

You can always reduce the 'amount' of data and go from ratio, to interval, to ordinal, to nominal.
You can also go from nominal to ordinal (e.g. alphebatizing), or from ordinal to interval to ratio, but those tend to be fairly arbitrary.

### Case study: Quantifying visual appeal
##### Background
"Infographic Aesthetics..."
Lots of new things: Tableau, R Studio...

Edward Tufte doesn't like "Chartjunk" - icons, funny shaped bars, etc.
On the other hand, Chartjunk increases data retention (Bateman et al., 2010.) Infographics are more memorable (Borkin et al., 2013)

##### Methodology
Picked 330 infographics and showed 30 them for 500 ms, then repeated with the same 30. Still online at http://www.labinthewild.org
##### Findings
- What makes an infographic appealing?
    - Positive correlation for colorfulness & appeal (p < 0.001)
    - Females prefer less compleximages
    - Older people prefered less complexity
    - More education = less color, less complexity
- How long does it take to form an impression?
- How consistent are people's judgements? (pretty consistent. People typically differed by less than 2/10 rates.)

- What factors make a website/infographic appealing?


### Using "Produce" type data: "Hindsight" study
##### Background
Visualizations that people are encountering are more complex.

There are really complex visualizations, which people just don't read that.

One proposal is keeping visualization history somehow (e.g. minimaps, purple links)

Different ways of keeping history:
- Augmentation (e.g. thickening an existing line)
- Addition (Add a new visual encoding e.g. heatmap in background...)
- Adaptation (when no visual channels available, modify an existing one)

##### Methodology
Darkened visited charts.
Paid Mechanical Turk people to look at charts and list 5 insights.

* Visits
* Revisits
* Exploration time

###### Results
- Average participants visits more sub-charts, and certainly more than the original average.
- Hindsight encourages people to visit different parts of the data.
- Slightly increases visit time

### Using exploration data to drive machine learning: "dis(tance) function"
Have a simple multidimensional dataset that you can try to graph in 2D space.

Expert user looks at 2D representation, and can say "These dots should be close together".
Then the computer changes the 2D view to match changes.

#### Just noticable differences (J-N-D)

Weber's law
$\Delta P = k * \frac{\Delta i}{i}$
Change in intensity = k * (change in intensity / total intensity)

###### stevens psychophysical  power law
We overestimate and underestimate different stimuli at different rates

Pre-attentive processing, hard-wired many-channels - it's easy to find certain differences if they're simple or unique. Useful for pointing out outliers and the like, but is very easy to mess up.
https://en.wikipedia.org/wiki/Pre-attentive_processing

##### Separable vs. Integral
Certain attributes (e.g. position and hue) don't mess each other up, but others (e.g. length and width) may mess it up more.
